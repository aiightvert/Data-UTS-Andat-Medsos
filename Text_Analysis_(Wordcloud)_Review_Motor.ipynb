# 3. Data Preprocessing



Mengubah kolom 'Comments' ke dalam bentuk string

```python
df['Comments'] = df['Comments'].astype(str)
```

Menghilangkan kolom 'Comments' yang kosong

```python
df = df[~df['Comments'].isnull()]
len(df)
```

Menghilangkan karakter-karakter spesial yang tidak dibutuhkan dari kolom 'Comments'

```python
def clean(txt):
    txt = txt.str.replace("()", "")
    txt = txt.str.replace('(<a).*(>).*()', '')
    txt = txt.str.replace('(&amp)', '')
    txt = txt.str.replace('(&gt)', '')
    txt = txt.str.replace('(&lt)', '')
    txt = txt.str.replace('(\xa0)', ' ')
    return txt

df['Comments'] = clean(df['Comments'])
```

Mengubah seluruh kata menjadi huruf kecil dalam kolom 'Comments'

```python
df['Comments'] = df['Comments'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df.head()
```

Menghilangkan tanda baca dari kolom 'Comments'

```python
df['Comments'] = df['Comments'].str.replace('[^\w\s]', '')
df.head()
```

Menghilangkan *stopwords* dari kolom 'Comments'

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('indonesian')
df['Comments'] = df['Comments'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df['Comments'].head()
```

Menghilangkan kata-kata yang hanya muncul sesekali saja (langka) pada kolom 'Comments'

```python
freq = pd.Series(' '.join(df['Comments']).split()).value_counts()
less_freq = list(freq[freq ==1].index)
less_freq
```

Mengkoreksi pengejaan masing-masing kata dalam kolom 'Comments'

```python
from textblob import TextBlob, Word, Blobber
df['Comments'].apply(lambda x: str(TextBlob(x).correct()))
df['Comments'].head()
```

*Stemming and Lemmatization* untuk mengubah masing-masing kata dari kolom 'Comments' ke dalam bentuk paling sederhana-nya tanpa imbuhan

```python
!pip install Sastrawi
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
df['Comments'] = df['Comments'].apply(lambda x: " ".join([stemmer.stem(word) for word in x.split()]))
df['Comments'].head()
```

Menghilangkan tanda baca dari kolom 'Comments' kembali

```python
df['Comments'] = df['Comments'].str.replace('[^\w\s]', '')
df['Comments'].head()
```

# 4. Data Analysis

Menambahkan kolom:
- Panjang_Comments : untuk menghitung jumlah string dalam masing-masing baris untuk kolom 'Comments'
- Jumlah_Kata : untuk menghitung jumlah kata dalam masing-masing baris untuk kolom 'Comments'



```python
df['Panjang_Comments'] = df['Comments'].astype(str).apply(len)
df['Jumlah_Kata'] = df['Comments'].apply(lambda x: len(str(x).split()))
df.head()
```

Menambahkan kolom 'Polarity' untuk menentukan nilai sentimen positif atau negatifnya Comments dalam range -1 sampai dengan 1.

```python
df['polarity'] = df['Comments'].map(lambda text: TextBlob(text).sentiment.polarity)
df.head()
```

```python
text = " ".join(review for review in df.Comments)

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

stopwords = set(stopwords.words('indonesian'))
wordcl = WordCloud(stopwords = stopwords, background_color='black', max_font_size = 50, max_words = 5000).generate(text)
plt.figure(figsize=(14, 12))
plt.imshow(wordcl, interpolation='bilinear')
plt.axis('off')
plt.show()
```
